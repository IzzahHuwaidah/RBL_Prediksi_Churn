### **Section 6: Data Preprocessing**
Sebelum melatih model, kita perlu menyiapkan datanya:
1. **Menghapus kolom yang tidak diperlukan** (customer_id)
2. **Melakukan encoding pada variabel kategorikal** (mengubah teks menjadi angka)
3. **Membagi data** menjadi data latih dan data uji
4. **Melakukan scaling pada fitur** (menormalisasi nilai numerik)


---

```python
print("=" * 70)
print("DATA PREPROCESSING (Praproses Data)")
print("=" * 70)

# Membuat salinan dataset agar data asli tetap aman
df_clean = df.copy()

# Langkah 1: Hapus customer_id (tidak berguna untuk prediksi)
df_clean = df_clean.drop('customer_id', axis=1)
print("\nLangkah 1: Kolom 'customer_id' dihapus")

# Langkah 2: Encoding Gender (Biner: Female=0, Male=1)
le_gender = LabelEncoder()
df_clean['gender'] = le_gender.fit_transform(df_clean['gender'])
print("Langkah 2: Melakukan encoding pada kolom 'gender' -> Female=0, Male=1")

# Langkah 3: One-Hot Encoding untuk Country (membuat kolom dummy)
# drop_first=True digunakan untuk menghindari multikolinearitas
df_clean = pd.get_dummies(df_clean, columns=['country'], prefix='country', drop_first=True)
print("Langkah 3: Melakukan one-hot encoding pada kolom 'country'")
print(f"   Kolom baru: {[col for col in df_clean.columns if 'country' in col]}")

# Menampilkan informasi dataset yang sudah dibersihkan
print(f"\nUkuran Dataset Setelah Pembersihan: {df_clean.shape}")
print(f"\nDaftar Kolom Fitur Akhir ({len(df_clean.columns)-1} fitur):")
for i, col in enumerate([c for c in df_clean.columns if c != 'churn'], 1):
    print(f"   {i}. {col}")
```

```python
print("\n" + "=" * 70)
print("PEMBAGIAN DATA (SPLITTING DATA)")
print("=" * 70)

# Memisahkan fitur (X) dan target (y)
X = df_clean.drop('churn', axis=1)
y = df_clean['churn']

print(f"\nFitur (X): {X.shape} → {X.shape[1]} fitur × {X.shape[0]} sampel")
print(f"Target (y): {y.shape} → {y.shape[0]} label")

# Membagi data menjadi train (80%) dan test (20%)
# stratify=y memastikan proporsi churn tetap sama di kedua set
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% untuk data testing
    random_state=42,    # Reproducible (hasil tetap)
    stratify=y          # Menjaga proporsi churn di train & test
)

print(f"\nData Training : {X_train.shape[0]} sampel ({X_train.shape[0]/len(df)*100:.1f}%)")
print(f"Data Testing  : {X_test.shape[0]} sampel ({X_test.shape[0]/len(df)*100:.1f}%)")

# Mengecek apakah stratifikasi berjalan dengan benar
print(f"\nPengecekan Distribusi Churn:")
print(f"   Training : {y_train.mean()*100:.2f}% churn")
print(f"   Testing  : {y_test.mean()*100:.2f}% churn")
print(f"   Original : {y.mean()*100:.2f}% churn (keseluruhan data)")
```

```python
print("\n" + "=" * 70)
print("FEATURE SCALING (Normalisasi Fitur)")
print("=" * 70)

# StandardScaler: mengubah fitur agar memiliki mean=0 dan std=1
# Ini membantu algoritma ML bekerja lebih baik dan konvergensi lebih cepat
scaler = StandardScaler()

# Fit hanya pada data training (untuk menghindari data leakage!)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nFitur telah dinormalisasi menggunakan StandardScaler")
print("   - Rata-rata (mean) ≈ 0")
print("   - Simpangan baku (standard deviation) ≈ 1")

# Contoh hasil scaling
print(f"\nContoh - Credit Score sebelum scaling:")
print(f"   Training: mean={X_train['credit_score'].mean():.2f}, std={X_train['credit_score'].std():.2f}")

print(f"\nContoh - Credit Score setelah scaling:")
print(f"   Training: mean={X_train_scaled[:, X.columns.get_loc('credit_score')].mean():.2f}, " +
      f"std={X_train_scaled[:, X.columns.get_loc('credit_score')].std():.2f}")

print("\nData sekarang siap digunakan untuk pemodelan Machine Learning!")
```

### Apa Arti dari Metrik-Metrik Ini?

**Interpretasi Confusion Matrix:**
- **True Negatives (TN)**: Prediksi "Tetap" yang benar
- **False Positives (FP)**: Diprediksi "Churn", tetapi sebenarnya tetap (Kesalahan Tipe I)
- **False Negatives (FN)**: Diprediksi "Tetap", tetapi sebenarnya churn (Kesalahan Tipe II)
- **True Positives (TP)**: Prediksi "Churn" yang benar

**Kesalahan Mana yang Lebih Buruk?**
- **False Negatives** biasanya lebih buruk bagi bank — karena bank gagal mendeteksi nasabah yang akan churn, menyebabkan kehilangan pendapatan
- **False Positives** mengakibatkan biaya tambahan untuk melakukan upaya retensi, tetapi dapat mencegah kehilangan pendapatan

```python
print("\n" + "=" * 70)
print("ANALISIS DAMPAK BISNIS")
print("=" * 70)

# Asumsi untuk perhitungan dampak bisnis
avg_customer_value = 50000       # Nilai rata-rata saldo nasabah
retention_cost = 500             # Biaya retensi per nasabah
retention_success_rate = 0.3     # 30% upaya retensi berhasil

# Hitung metrik bisnis
tn, fp, fn, tp = cm.ravel()

# Biaya dan Pendapatan
saved_customers = int(tp * retention_success_rate)
retention_spending = (tp + fp) * retention_cost
revenue_saved = saved_customers * avg_customer_value
lost_revenue = fn * avg_customer_value
net_benefit = revenue_saved - retention_spending

print(f"\nMetrik Bisnis (Berdasarkan Data Test):")
print(f"\nDampak Pendapatan:")
print(f"   Jumlah nasabah yang terdeteksi akan churn: {tp}")
print(f"   Potensi nasabah yang diselamatkan (30% berhasil): {saved_customers}")
print(f"   Pendapatan yang terselamatkan: Rp{revenue_saved:,.0f}")
print(f"   Pendapatan hilang (nasabah churn yang tidak terdeteksi): Rp{lost_revenue:,.0f}")

print(f"\nAnalisis Biaya:")
print(f"   Jumlah kampanye retensi dikirim: {tp + fp}")
print(f"   Total biaya retensi: Rp{retention_spending:,.0f}")

print(f"\nDampak Bisnis Bersih:")
print(f"   Keuntungan Bersih (Net Benefit): Rp{net_benefit:,.0f}")
print(f"   ROI: {(net_benefit/retention_spending)*100:.1f}%")

# Visualisasi dampak bisnis
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Breakdown dampak bisnis
impact_categories = ['Pendapatan Terselamatkan', 'Pendapatan Hilang', 'Biaya Retensi']
impact_values = [revenue_saved, -lost_revenue, -retention_spending]
impact_colors = ['#2ecc71', '#e74c3c', '#f39c12']

axes[0].bar(impact_categories, impact_values, color=impact_colors, 
            edgecolor='black', linewidth=2, alpha=0.8)
axes[0].axhline(y=0, color='black', linewidth=2)
axes[0].set_ylabel('Jumlah (Rp)', fontsize=12, fontweight='bold')
axes[0].set_title('Rincian Dampak Bisnis', fontsize=14, fontweight='bold', pad=15)
axes[0].grid(axis='y', alpha=0.3)

for i, v in enumerate(impact_values):
    axes[0].text(i, v + (1000 if v > 0 else -5000), 
                 f'Rp{abs(v):,.0f}', ha='center', 
                 fontweight='bold', fontsize=10)

# Visualisasi nilai bersih (dengan vs tanpa model)
categories = ['Tanpa Model', 'Dengan Model']
without_model = -lost_revenue - fn * retention_cost   # Semua churn hilang
with_model = net_benefit

axes[1].bar(categories, [without_model, with_model], 
            color=['#e74c3c', '#2ecc71'], 
            edgecolor='black', linewidth=2, alpha=0.8)
axes[1].axhline(y=0, color='black', linewidth=2)
axes[1].set_ylabel('Keuntungan Bersih (Rp)', fontsize=12, fontweight='bold')
axes[1].set_title('Perbandingan Nilai Model', fontsize=14, fontweight='bold', pad=15)
axes[1].grid(axis='y', alpha=0.3)

for i, v in enumerate([without_model, with_model]):
    axes[1].text(i, v + (2000 if v > 0 else -8000), 
                 f'Rp{v:,.0f}', ha='center', 
                 fontweight='bold', fontsize=11)

plt.tight_layout()
plt.savefig('9_business_impact.png', dpi=300, bbox_inches='tight')
plt.show()
```
